# unión de datos

```python
df_clim = pd.read_csv('New_DB_2.csv')
data_full = pd.read_csv('data_full_process.csv')
valor = data_full[['Latitud', 'Longitud', 'Valor']]
unique_combinations_df1 = df_clim.drop_duplicates(subset=['Longitud', 'Latitud', 'Fecha Evento']).shape[0]
unique_combinations_df2 = data_full.drop_duplicates(subset=['Longitud', 'Latitud', 'Fecha Evento']).shape[0]

print(f"Combinaciones únicas en df1: {unique_combinations_df1}")
print(f"Combinaciones únicas en df2: {unique_combinations_df2}")

merged_df = pd.merge(df_clim, valor, on=['Latitud', 'Longitud'], how='left')
merged_df = merged_df.drop_duplicates()
merged_df = merged_df.reset_index()
merged_df
```

```python
print(pd.read_csv('tif_data').shape)
print(pd.read_csv('tif_data_0_01').shape)
tif_data = pd.read_csv('tif_data')
tif_data_0_01 = pd.read_csv('tif_data_0_01')
print(tif_data.shape, tif_data_0_01.shape)
tif_data_0_01 = tif_data_0_01.drop(columns=['Unnamed: 0'])
tif_data = tif_data.drop(columns=['Unnamed: 0'])
tif_data['Valor'] = 1
tif_data_0_01['Valor'] = 0
```

```python
tif_concatenate = pd.concat([tif_data, tif_data_0_01],)
tif_concatenate.shape
```

```python
cols = ['slope', 'PP', 'valor_humedad_suelo1','valor_humedad_suelo2', 'valor_humedad_suelo3', 'valor_humedad_suelo4']
merged_df[cols] = merged_df[cols].fillna(merged_df[cols].mean())
```

```python
tif_merged = pd.merge(tif_concatenate, merged_df, on=['Longitud', 'Latitud','Valor'],how='inner')
```

```python
tif_merged = tif_merged.drop_duplicates(subset=['Longitud', 'Latitud', 'Fecha Evento_x'])
tif_merged = tif_merged.reset_index()
```

```python
tif_merged.shape
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
sns.set_palette("pastel")
plt.figure(figsize=(8, 6))
g = sns.countplot(tif_merged, x='Valor', color='skyblue')

# Obtener la cantidad de 1s y 0s en la columna 'Valor'
value_counts = tif_merged['Valor'].value_counts()

# Añadir etiquetas al gráfico para indicar la cantidad de 1s y 0s
plt.text(0, value_counts[0], f'{value_counts[0]}', ha='center', va='bottom')
plt.text(1, value_counts[1], f'{value_counts[1]}', ha='center', va='bottom')
# Añadir título y etiquetas de los ejes
plt.title('Histograma de valores')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

# Mostrar el gráfico
plt.show()
```
