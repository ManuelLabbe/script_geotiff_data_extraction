```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_corr_map(df: pd.DataFrame):
    plt.figure(figsize=(15,14))
    heatmap = sns.heatmap(df.corr(), annot=True, cmap='pink_r')
    heatmap.set_title('Mapa de correlación')
    plt.show()
```

```python
# extración de la base de datos nueva
coso = pd.read_csv('coso.csv')
new0 = pd.read_csv('New_DB_new0.csv')
new0 = new0[['PP', 'slope', 'valor_humedad_suelo1', 'Latitud', 'Longitud']].dropna()
coso = coso[['PP', 'slope', 'valor_humedad_suelo1', 'Latitud', 'Longitud']].dropna()
data_0 = pd.read_csv('xterrae_data_0').dropna()
data_1 = pd.read_csv('xterrae_data_1').dropna()
data_0 = pd.merge(new0, data_0, on=['Latitud', 'Longitud'], how='inner')
data_1 = pd.merge(coso, data_1, on=['Latitud', 'Longitud'], how='inner')
data_0 = data_0.drop(columns=['Unnamed: 0', 'Latitud', 'Longitud'])
data_1 = data_1.drop(columns=['Unnamed: 0', 'Latitud', 'Longitud'])
data_0['Valor'] = 0
data_1['Valor'] = 1
data = pd.concat([data_0, data_1])
```

```python
df_filt = data.filter(like='PIRange', axis=1)
df_filt
```

```python
plot_corr_map(df_filt)
```

```python
rosetta_mean_1 = pd.read_csv('rosetta_mean_1')
rosetta_mean_1 = rosetta_mean_1.drop(columns=['Unnamed: 0.1', 'Unnamed: 0', 'Latitud', 'Longitud'])
```

```python
plot_corr_map(data[rosetta_mean_1.columns])
```

```python
soilmaps_mean_1 = pd.read_csv('soilmaps_mean_1')
soilmaps_mean_1 = soilmaps_mean_1.drop(columns=['Unnamed: 0.1', 'Unnamed: 0','Latitud', 'Longitud'])
```

```python
plot_corr_map(data[soilmaps_mean_1.columns])
```
